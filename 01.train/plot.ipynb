```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("lcurve.out", names=True)
for name in data.dtype.names[1:-1]:
    plt.plot(data['step'], data[name], label=name)
plt.legend()
plt.xlabel('Step')
plt.ylabel('Loss')
# plt.xscale('log')
plt.yscale('log')
plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9,6.66))
plt.rc('font',family='Arial', weight='medium')
ax.set_facecolor('white')

kuang = ['top','bottom','left','right']
for i in kuang:
    ax.spines[i].set_color('black')
    ax.spines[i].set_linewidth(1.5)

for name in data.dtype.names[1:-1]:
    ax.plot(data['step'], data[name], label=name)

lgd = ax.legend(fontsize=16)
lgd.get_frame().set_alpha(0.1)

ax.set_xlabel('Step', fontsize=16)
ax.set_ylabel('Loss', fontsize=16)
ax.tick_params(axis='both', labelsize=16)
plt.yscale('log')
plt.grid()

_ = plt.savefig('./fig/Loss.png',dpi=360, bbox_inches = 'tight')
```
